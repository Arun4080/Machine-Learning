# Linear Regression in Detail

### So whole linear Regression is based on drawing a best fit line with given data, here we use old formula whilch is 
###  Y=m(X) + C 
### and for Slope we have 
### m(Slope) =(X'.Y'-(X.Y)') / ((X')^2 - (X^2)') 

## Importing Libraries

```python
from statistics import mean
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
```

### first we must have some short of data to do Linear Regression so we will take 2 arrays to keep doing our work

```python
xs=np.array([3,2,3,4,8,6], dtype= np.float64)
ys=np.array([3,4,2,5,6,9], dtype= np.float64)
```

###    Now lets make a Function to find best fit slope using upper formula --    (m(Slope)) and value of C

```python
def best_fit_slope_and_intercept(xs,ys):
    m = ( ((mean(xs)*mean(ys)) - mean(xs*ys)) /
         ((mean(xs)*mean(xs)) - mean(xs*xs)) )
    c = mean(ys) - m*mean(xs)
    return m, c
```

### lets find Regression line cordinates

```python
m,c = best_fit_slope_and_intercept(xs, ys)

regression_line = [(m*x)+c for x in xs]
```

## Lets draw Regression line

```python
style.use('fivethirtyeight')
plt.scatter(xs,ys)
plt.plot(xs, regression_line)
plt.show()
```

### lets predict some data

```python
predict_x = 10
predict_y= (m*predict_x)+c

plt.scatter(xs,ys)
plt.scatter(predict_x,predict_y, color='r')
plt.plot(xs, regression_line)
plt.show()
```
